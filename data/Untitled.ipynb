```python
import json
import pandas as pd
```

```python
json_file_path = "scatter_cartogram.json"

with open(json_file_path, 'r') as j:
     contents = json.loads(j.read())
```

```python
contents
```

```python
list(contents.keys())
```

```python
d2016 = pd.read_excel('2016_data.xlsx')
d2017 = pd.read_excel('2017_data.xlsx')
d2018 = pd.read_excel('2018_data.xlsx')
d2019 = pd.read_excel('2019_data.xlsx')
d2020 = pd.read_excel('2020_data.xlsx')
d2021 = pd.read_excel('2021_data.xlsx')
d2022 = pd.read_excel('2022_data.xlsx')

data_list = [d2016, d2017, d2018, d2019, d2020, d2021, d2022]
```

```python
oil_price = pd.read_excel('oil_price.xlsx')
oil_price
```

```python
list(oil_price)
```

```python
oil_price[oil_price['State'] == 'Alabama']
```

```python
min_value = 1e20
max_value = 0
for state in list(contents.keys()):
    for data_index, data in enumerate(data_list):
        state_info = contents[state]
#         print('state:', state)
        
        value1 = list(data[data['State'] == state]['Electric (EV)'])[0]
        value2 = list(data[data['State'] == state]['Plug-In Hybrid Electric (PHEV)'])[0]
        value3 = list(data[data['State'] == state]['Hybrid Electric (HEV)'])[0]
        
        value = value1 + value2 + value3
        
        if 'EV_data' not in state_info:
            state_info['EV_data'] = [value]
        else:
            state_info['EV_data'].append(value)
            
        min_value = min(min_value, value)
        max_value = max(max_value, value)
        
        state_info['EV'] = str(value)
        state_info['scatter']['Gasoline'] = str(list(data[data['State'] == state]['Gasoline'])[0])
        state_info['scatter']['oil_price'] = list(oil_price[oil_price['State'] == state]['Regular'])[0]
```

```python
print(min_value)
print(max_value)
```
